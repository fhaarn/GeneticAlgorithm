```
#Muhammad Farhan Audianto || 1301180466 || IF42INT
import random
import numpy as np


#generate random chromosome
def ch(nKrom):
  chr = []
  for i in range(nKrom):
    chr.append(random.randint(0,1))
  return chr

#generate random population
def pop(nKrom,nPop) :
  popu = []
  for i in range(nPop):
    popu.append(ch(nKrom))
  return popu

#search x1 x2
def form(half,flag) :
  #boundary base on the question
  b1 = [-1,2]
  b2 = [-1,1]
  denominator = 0
  for i in range(len(half)):
    denominator=denominator+2**(-(i+1))
  rt = 0
  for i in range(len(half)):
    rt = rt + half[i] * 2 ** (-(i+1))
  if flag == 1 :
    return b1[0] + ((b1[1]-b1[0])/denominator)*rt
  else :
    return  b2[0] + ((b2[1]-b2[0])/denominator)*rt


#dividing the chromosome
def decode(ind):
  x1 = form(ind[:len(ind)//2],1)
  x2 = form(ind[len(ind)//2:],2)
  return x1,x2

#search h
def h (x1,x2) :
  return (np.cos(x1) * np.sin(x2)) - (x1 / ((x2**2) + 1))

#search f
def f (h) :
  return -h

#evaluate all of the chromosom f value
def evaluate(population) :
  grpop = []
  for i in range(len(population)) :
    x1,x2 = decode(population[i])
    grpop.append(f(h(x1,x2)))
  return grpop 

#elitism
def elitism(pop,fit) :
  popandfit = sorted(zip(pop,fit), key=lambda x: x[1],reverse=True )
  eli = []
  for i in range(5):
    eli.append(pop[i][0])
  return [eli]

#search parent
def parentSelection (pop):
  par = random.sample(pop,k=5)
  fitpar = evaluate(par)
  popandfit = sorted(zip(par,fitpar), key=lambda x: x[1],reverse=True )
  return popandfit[0][0]

#do crossover
def crossover(parent1,parent2):
  if random.random()<0.7 :
    p = random.randint(0,len(parent1))
    return parent1[:p] + parent2[p:]
  
  return parent1

#do mutation
def mutation(offspring) :
  if random.random()<0.1 :
    p = random.randint(0,len(offspring)-1)
    q = random.randint(0,len(offspring)-1)
    if p == q :
      while p == q :
        q = random.randint(0,len(offspring)-1)
    offspring[p] = random.randint(0,1)
    offspring[q] = random.randint(0,1)
  return offspring

def main() :
  nGeneration = 10
  population = pop(8,10)
  for i in range(nGeneration):
    fit = evaluate(population)
    newPop = elitism(population,fit)
    while len(newPop) <= len(population) :
      p1 = parentSelection(population)
      p2 = parentSelection(population)
      child = crossover(p1,p2)
      child = mutation(child)
      newPop.append(child)
    population = newPop.copy()
  return list(sorted(zip(population,fit), key=lambda x: x[1],reverse=True ))
```
